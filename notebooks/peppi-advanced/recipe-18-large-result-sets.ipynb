# Recipe 18: Handle Large Result Sets Efficiently

Process thousands of products without running out of memory.

```python
import pds.peppi as pep
```

```python
client = pep.PDSRegistryClient()

# Get only the fields we need for efficiency
fields = ['lid', 'pds:Identification_Area.pds:title']

products = pep.Products(client) \
    .has_target("Mars") \
    .observationals() \
    .fields(fields)

# Process in batches
batch_size = 100
processed = 0

batch = []
for product in products:
    batch.append(product.id)
    
    if len(batch) >= batch_size:
        print(f"Processing batch: products {processed} to {processed + len(batch)}...")
        # Process this batch (e.g., save_to_database(batch))
        
        batch = []
        processed += batch_size
    
    if processed >= 500:  # Limit for demo
        break

print(f"Total processed: {processed}  + {len(batch)} products")
```
