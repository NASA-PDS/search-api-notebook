```python
from __future__ import print_function
import pds.api_client as pds_api
from pprint import pprint
```

```python
configuration = pds_api.Configuration()

# production server
configuration.host = 'https://pds.nasa.gov/api/search/1.0'

api_client = pds_api.ApiClient(configuration)
```

## Create an instance of the API class

```python
from pds.api_client.api.bundles_api import BundlesApi
bundles = BundlesApi(api_client)

bundles_response = bundles.get_bundles(q='lidvid eq "urn:nasa:pds:insight_documents::2.0"', 
                                       fields=["lidvid"],
                                        _content_type='application/kvp+json'
                                      )
```

```python
from pds.api_client.api.collections_api import CollectionsApi
api_instance = CollectionsApi(api_client)

```

## Request the API

```python
try:
    # capabilities api entry point, list the resources provided by the current API end-point.
    api_response = api_instance.get_collection(q="", start=0, limit=20)
    pprint(api_response)
except ApiException as e:
    print("Exception when calling CollectionsApi->get_collection: %s\n" % e)
```
