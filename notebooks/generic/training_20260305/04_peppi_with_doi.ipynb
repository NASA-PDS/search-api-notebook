# Get data with given DOI with peppi

```python
import pds.peppi as peppi
from pprint import pprint
```

```python
client = peppi.PDSRegistryClient()

products = peppi.Products(client)

# with_doi() was removed in pds.peppi 0.8.x; use filter() with the DOI field directly
my_products = products.filter('pds:Citation_Information.pds:doi eq "10.17189/1522910"')
```

```python
for p in my_products:
    pprint(p)
```
