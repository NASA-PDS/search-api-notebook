# Search data with a given DOI

Here using the doi 10.17189/1522910

```python
import requests
from pprint import pprint
```

## Buid my request

```python
q_string = '(pds:Citation_Information.pds:doi eq "10.17189/1522910")'

params = {
    'q': q_string,
    'fields': ",".join(["lidvid", "pds:Citation_Information.pds:doi"]),
    'limit': 3000
}

headers = {
    'Accept': 'application/json'
}

response = requests.get(
    'https://pds.nasa.gov/api/search/1/products',
    headers=headers,
    params=params
)

response = response.json()
```
