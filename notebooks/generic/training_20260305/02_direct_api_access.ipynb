# Import the libraries

We rely on the HTTP RestFul client library `requests`

```python
import requests
from pprint import pprint
```

# Make your selection

Define your query parameters as described in the [documentation](https://nasa-pds.github.io/pds-api/guides/search/endpoints.html#query-parameters)

The server name (https://pds.nasa.gov/api/search/1/products) is explicitly defined but you don't need to change it.

```python
q_string = (
    '('
       '(pds:Science_Facets.pds:wavelength_range eq "Visible") '
       'and (pds:Science_Facets.pds:wavelength_range eq "Visible") '
       'and (pds:Header.pds:parsing_standard_id eq "FITS 3.0") '
       'and (pds:Time_Coordinates.pds:start_date_time gt "2022-04-30T00:00:00Z")'
    ')'
)

params = {
    'q': q_string,
    'fields': ",".join(["lidvid", "pds:Science_Facets.pds:wavelength_range"]),
    'limit': 3000
}

headers = {
    'Accept': 'application/json'
}

response = requests.get(
    'https://pds.nasa.gov/api/search/1/products',
    headers=headers,
    params=params
)

response = response.json()
```

# Use the extracted response

See the number of results

```python
response["summary"]["hits"]
```

Iterate on the results

```python
for i, product in enumerate(response["data"]):
    pprint(product)

    if i>3:
        break
    
```
