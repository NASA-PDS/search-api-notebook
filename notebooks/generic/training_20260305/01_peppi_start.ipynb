# Simple Peppi Template

## Import the required libraries

```python
import pds.peppi as peppi
from datetime import datetime
from pprint import pprint

```

## Connect to the API

```python
client = peppi.PDSRegistryClient()
```

## Build your query

Start from the template and add your own selection criteria using the methods proposed in the [reference documentation](https://nasa-pds.github.io/peppi/reference.html#pds.peppi.query_builder.QueryBuilder).

```python
products = peppi.Products(client)

# Add your criterias, use TAB for autocompletion
# the criteria add up with AND operator
# for example
my_products = products.of_collection("urn:nasa:pds:mess-rs-raw:data-rsr::1.0")

# add another critiria on top of your selection
end_date = datetime.fromisoformat("2012-03-05")
my_products = my_products.before(end_date)

# Make it your own...
```

## Request the API, get your metadata/data

You have 2 options:
- loop on the results, pagination is handle by peppi for you
- get a table of results

### Loop on the results

```python
for i, product in enumerate(my_products):
    # Do something 
    pprint(product)
    
    # here we leave the loop
    if i>2:
        break
```

### Get a table of all products

Using pandas DataFrame export.

```python
my_products = my_products.fields(["lidvid", "ops:Data_File_Info.ops:file_size"])

import pandas as pd

df = my_products.as_dataframe(max_rows=10)

df
```
